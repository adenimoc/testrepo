Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
# Add marker_cluster to current site_map
site_map.add_child(marker_cluster)

for index, record in spacex_df.iterrows():
    # Crear contenido para el popup
    outcome = "Success" if record['class'] == 1 else "Failure"
    popup_content = f"""
    <b>Launch Site:</b> {record['Launch Site']}<br>
    <b>Outcome:</b> {outcome}<br>
    <b>Coordinates:</b> {record['Lat']:.4f}, {record['Long']:.4f}
    """
    
    # Crear objeto Marker con personalización
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],
        popup=folium.Popup(popup_content, max_width=250),
        icon=folium.Icon(
            color='white',              # Color de fondo del ícono
            icon_color=record['marker_color'],  # Color principal (verde/rojo)
            icon='rocket' if record['class'] == 1 else 'remove-sign',
            prefix='glyphicon'  # Usar glyphicons de Bootstrap
        )
    )
    
    # Agregar el marcador al cluster
    marker_cluster.add_child(marker)

# 4. Agregar el cluster al mapa (CORRECCIÓN CLAVE)
site_map.add_child(marker_cluster)  # Esta línea estaba incompleta en el código original

# 5. Opcional: Agregar control de capas
folium.LayerControl().add_to(site_map)

# Mostrar el mapa
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
coastline_coords = {
    'CCAFS LC-40': (28.56367, -80.57163),
    'CCAFS SLC-40': (28.56319, -80.56798),
    'KSC LC-39A': (28.58010, -80.64216),
    'VAFB SLC-4E': (34.63474, -120.62516)
}
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
# distance_marker = folium.Marker(
#    coordinate,
#    icon=DivIcon(
#        icon_size=(20,20),
#        icon_anchor=(0,0),
#        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
#        )
#    )
# Calcular distancias y agregar marcadores
for site_name, site_info in spacex_df.iterrows():
    site_lat = site_info['Lat']
    site_lon = site_info['Long']
    site_name = site_info['Launch Site']
    
    if site_name in coastline_coords:
        coast_lat, coast_lon = coastline_coords[site_name]
        
        # Calcular distancia
        distance = calculate_distance(site_lat, site_lon, coast_lat, coast_lon)
        
        # Crear marcador en la costa con la distancia
        distance_marker = folium.Marker(
            [coast_lat, coast_lon],
            icon=DivIcon(
                icon_size=(150,40),  # Ajustado para mejor visualización
                icon_anchor=(75,20),  # Centrar el texto
                html=f'<div style="font-size: 12; color:#d35400;"><b>{distance:.2f} KM</b></div>'
            )
        )
        distance_marker.add_to(site_map)
        
        # Agregar línea entre sitio y costa
        line = folium.PolyLine(
            locations=[[site_lat, site_lon], [coast_lat, coast_lon]],
            color='blue',
            weight=1
        )
        line.add_to(site_map)
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
```

```python
# Agregar posición del mouse al mapa
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)
site_map.add_child(mouse_position)

# Función para calcular distancia
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # Radio de la Tierra en km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    return R * c

# Definir puntos de interés para cada sitio
proximities = {
    'CCAFS LC-40': {
        'city': ('Titusville', 28.612, -80.807),
        'railway': ('Ferrocarril', 28.57208, -80.58528),
        'highway': ('Carretera', 28.56215, -80.57056),
        'coastline': ('Costa', 28.56367, -80.57163)
    },
    'CCAFS SLC-40': {
        'city': ('Titusville', 28.612, -80.807),
        'railway': ('Ferrocarril', 28.57208, -80.58528),
        'highway': ('Carretera', 28.56335, -80.57081),
        'coastline': ('Costa', 28.56319, -80.56798)
    },
    'KSC LC-39A': {
        'city': ('Titusville', 28.612, -80.807),
        'railway': ('Ferrocarril', 28.57208, -80.58528),
        'highway': ('Carretera', 28.57355, -80.65505),
        'coastline': ('Costa', 28.58010, -80.64216)
    },
    'VAFB SLC-4E': {
        'city': ('Lompoc', 34.638, -120.458),
        'railway': ('Ferrocarril', 34.636, -120.624),
        'highway': ('Carretera', 34.739, -120.582),
        'coastline': ('Costa', 34.63474, -120.62516)
    }
}

# Colores para diferentes tipos de proximidad
proximity_colors = {
    'city': 'green',
    'railway': 'brown',
    'highway': 'gray',
    'coastline': 'blue'
}

# Procesar cada sitio de lanzamiento
for site_name, site_info in spacex_df.iterrows():
    site_lat = site_info['Lat']
    site_lon = site_info['Long']
    site_name = site_info['Launch Site']
    
    if site_name in proximities:
        site_prox = proximities[site_name]
        
        # Procesar cada tipo de proximidad
        for prox_type, prox_data in site_prox.items():
            prox_name, prox_lat, prox_lon = prox_data
            
            # Calcular distancia
            distance = calculate_distance(site_lat, site_lon, prox_lat, prox_lon)
            
            # Crear marcador con distancia
            marker = folium.Marker(
                [prox_lat, prox_lon],
                icon=DivIcon(
                    icon_size=(150,40),
                    icon_anchor=(75,20),
                    html=f'<div style="font-size: 12; color:{proximity_colors[prox_type]}"><b>{prox_name}: {distance:.2f} KM</b></div>'
                )
            )
            marker.add_to(site_map)
            
            # Dibujar línea entre sitio y proximidad
            line = folium.PolyLine(
                locations=[[site_lat, site_lon], [prox_lat, prox_lon]],
                color=proximity_colors[prox_type],
                weight=1.5,
                dash_array='5,5' if prox_type == 'city' else None
            )
            line.add_to(site_map)

# Mostrar mapa actualizado
site_map
```
